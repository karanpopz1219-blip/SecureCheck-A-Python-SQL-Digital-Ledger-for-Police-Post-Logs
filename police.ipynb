```python
import pandas as pd

# Read the Excel file
file_path = "traffic_stops.csv"  
df = pd.read_csv(file_path)

# Display first few rows
df.head()
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Backward fill (bfill)
df_bfill = df.fillna(method='bfill')

# Forward fill (ffill)
df_ffill = df.fillna(method='ffill')

# Fill with mean (for numerical columns only)
df_mean = df.fillna(df.mean(numeric_only=True))

# Fill with mode (for categorical columns)
df_mode = df.fillna(df.mode().iloc[0])

# Fill with 0
df_zero = df.fillna(0)

# Drop rows with null values
df_dropped = df.dropna()

# Display results
print("Null values after using bfill:\n", df_bfill.isnull().sum())
print("Null values after using mean:\n", df_mean.isnull().sum())
print("Null values after using mode:\n", df_mode.isnull().sum())
print("Null values after using 0:\n", df_zero.isnull().sum())
print("Shape after dropping rows with null values:", df_dropped.shape)
```

```python
# Check actual column names
print(df.columns)

# Calculate mean of 'search_type' for USA and India
usa_india_mean_search_type = df[df["country_name"].isin(["usa", "india"])]["search_type"].mean()

# Fill NaN values in the 'search_type' column for USA and India
df.loc[(df["country_name"].isin(["usa", "india"])) & (df["search_type"].isnull()), "search_type"] = usa_india_mean_search_type

# Check if null values remain in USA and India 'search_type' column
print(df[df["country_name"].isin(["usa", "india"])]["search_type"].isnull().sum())
```
